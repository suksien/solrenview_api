```python
import pandas as pd
import numpy as np
```

```python
met_filename = 'MET 2025-10-24'
```

```python
file = open(met_filename, 'r')
```

```python
arr = []
for line in file:
    arr.append(line)
```

```python
# header is line 2 and data starts line 5
header = [string.split('"')[1] for string in arr[1].split(',')]
data = arr[4:]
real_data = [d.split(',') for d in data]
```

```python
df = pd.DataFrame(real_data, columns=header)
#df = df.replace('NAN', '')
for col in df.columns:
    df[col].replace('"NAN"', "", inplace=True)
```
